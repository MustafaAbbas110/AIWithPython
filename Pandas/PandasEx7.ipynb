```python
fake_company = {
                "Year":[1990,1990,1990,1991,1992,1992] ,
                "Name":["Alice","Bob","Charlie","Alice","Bob","Charlie"],
                "Department":["HR","RD","Admin","HR","RD","Admin"],
                "Age":[25,30,45,26,31,46],
                "Salary":[50000,48000,55000,52000,50000,60000]
               }
```

```python
data_company = pd.DataFrame(fake_company)
```

# Group By

```python
# Sum the Salary of Each Year
data_company.groupby(["Year"])["Salary"].sum()
```

```python
# Mean The Salary of Each Year
data_company.groupby(["Year"])["Salary"].mean()
```

```python
# Total Salary By Employee
data_company.groupby(["Name"])["Salary"].sum()
```

```python
# We display the salary distribution per department per year.
data.groupby(['Year', 'Department'])['Salary'].sum()
```
